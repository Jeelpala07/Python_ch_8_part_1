```python
## Line Plot

import matplotlib.pyplot as plt

x=[1,2,6,8]
y=[3,8,1,10]

plt.plot(x,y,marker="o",ms=20,mec='c',mfc='w',linestyle="-.",linewidth=5,color='k')
plt.show()

# ms-marker sizw 
# mec -marker edge color 
# mfc marker face color
# linestyle='dotted : ',solid - , double dash-- ,dash .
```

```python
x1=[1,2,3]
y1=[5,6,7]
x2=[3,2,5]
y2=[6,4,7]
plt.plot(x1,y1,marker="o",ms=20,mec='c',mfc='w',linestyle="-",linewidth=5,color='k')
plt.show()
plt.plot(x2,y2,marker="*",ms=20,mec='c',mfc='w',linestyle="--",linewidth=5,color='b')
plt.show()
```

```python
# labels tittle grid
y=[3,8,1,10]
plt.plot(y)
font={'family':'serif','color':'blue','size':20}
plt.title('line plot',fontdict=font,loc='left')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(axis='x',linestyle='--',linewidth=5)
plt.grid(axis='y',linestyle='--',linewidth=5)
plt.show()
```

```python
x=[5,7,8,7,2,17,2]
y=[99,68,87,88,111,86,103]
color=['red','blue','green','pink','black','m','y']
ss=[2020,3151,4542,1553,4564,1575,1056]
plt.scatter(x,y,c=y,s=ss,alpha=0.5,cmap='YlGn')
plt.colorbar()
plt.show()
```

```python
#  BAR CHART

x=['A','B','C','D']
y=[3,8,10,2]
plt.bar(x,y,color='r',width=0.8)
plt.show()
```

```python
x=['A','B','C','D']
y=[3,8,10,2]
plt.barh(x,y,color='r',height=0.8)
plt.show()
```

```python
x=['A','B','C','D']
y=[3,8,10,2]
a=['k','F','G','H']
b=[4,2,3,2]
plt.bar(x,y,color='r',width=0.8)
plt.bar(a,b,color='b',width=0.8)
plt.show()
```

```python
# HIstrogram
import matplotlib.pyplot as plt
a=[22,56,21,78,98,56,45,14,56,35,59,77,7,65,96]
plt.hist(a,bins=[0,25,50,75,100],histtype='bar',align='right',orientation='vertical',color='g',lw='5',ec='g')
```

```python
import matplotlib.pyplot as plt
a=[22,56,21,78,98,56,45,14,56,35,59,77,7,65,96]
plt.hist(a,bins=[0,25,50,75,100],histtype='step',align='left',orientation='vertical',color='g',lw='5',ec='g')
plt.xticks([0,25,50,75,100])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
y=[35,25,25,15]
plt.pie(y,labels=['a','b','c','d'],startangle=90,explode=[0.1,0,0,0],shadow=True,autopct='%1.f%%')
plt.legend(title='Pie Chart',loc=0)
```
